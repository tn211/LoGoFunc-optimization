```python
import pandas as pd
import json, os
```

### This decodes the ranked feature names.

```python
# Load the metrics CSV.
metrics = pd.read_csv('../metrics/ranked_by_accuracy.csv')
```

```python
# Ensure there are the correct number of metrics.
acc_ranks = metrics['Model'].tolist()
print(len(metrics))
```

```python
print(sorted(acc_ranks))
```

```python
# Ensure there are the correct number of rankings.
print(len(acc_ranks))
```

```python
path = '../metrics/accuracy_rankings.json'

acc_dict = {}

for index, ranking in enumerate(acc_ranks, start=1):
    acc_dict[index] = ranking

# Export dictionary to JSON.
with open(path, 'w') as json_file:
    json.dump(acc_dict, json_file, indent=4)

print("Saved to 'accuracy_rankings.json'")
```
