```python
import pandas as pd
import numpy as np
import json
```

### This just generates a list of the original features complete with numbering.

```python
df = pd.read_csv('../data/X_train_id.csv')
```

```python
print(df.shape)
```

```python
list = [f"{i+1}. {col}" for i, col in enumerate(df.columns)]
list
```

```python
# Place the columns inside a dictionary and export as JSON.
dict = {str(i+1): col for i, col in enumerate(df.columns)}
with open("../data/original_unranked_features.json", "w") as f:
    json.dump(dict, f, indent=4)
```

```python
# Place the columns inside a dictionary without ID and export as JSON.
no_id = df.drop(columns=['ID'], errors='ignore')
dict_no_id = {str(i+1): col for i, col in enumerate(no_id.columns)}
with open("../data/original_unranked_features.json", "w") as f:
    json.dump(dict_no_id, f, indent=4)
```
